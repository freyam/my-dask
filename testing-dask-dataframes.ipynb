```python
import dask.dataframe as dd

df = dd.demo.make_timeseries(
    start="2000",
    end="2015",
    dtypes={"A": float, "B": int, "C": int, "D": int, "E": str, "F": float, "G": float, "H": str, "I": str},
    freq="5s",
    partition_freq="3M",
    seed=1234,
)
df = df.shuffle(1000)


df.dask.visualize(color="layer_type", filename="7974-ZZ.png")
```

```python
import dask
import dask.dataframe as dd

df = dask.datasets.timeseries()
df2 = df[df.y > 0]
df3 = df2.groupby('name').x.std()

df3.dask.visualize(color="layer_type")
```
